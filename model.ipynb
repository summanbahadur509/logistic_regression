## Model

Logistic Regression is an algorithm that performs binary classification by modeling a dependent variable (Y) in terms of one or more independent variables (X). In other words, it’s a generalized model that predicts the probability that an event will occur.

```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from matplotlib import pyplot
```

```python
df = pd.read_spss('final_dataframe.sav')
```

```python
df.shape
pd.set_option('display.max_rows',100)
```

```python
df.head(100)
```

```python
df.columns = ['women_education','engaged_activity_for_income','help_in_family_business_past_week','estimation_overall_happiness','wanted_child_later','age_at_marriage','area','father_education','wanted_child_later_or_not','religious_influence']
```

There are eight input variables: examples include women education level, age at the time of marriage, father education level, area, estimation of overall happiness, religious influence etc. There is one output variable which describes whether or not the women agreed to have a child late or did not wanted at all. This binary output variable has two possible values: “yes” and “no”. 

```python
df.wanted_child_later.replace({'NONE / NO MORE':'NO','LATER':'YES'},inplace=True)
```

```python
df.area.replace({'RURAL':'NO','URBAN':'YES'},inplace=True)
```

Next we’ll check our dataset for null values using seaborn:

```python
sns.heatmap(df.isnull()) #Showing there are no null values
```

```python
#Count plot of output variable UN2 - Fertility preference variable - wanted child later or not
sns.countplot(x='wanted_child_later_or_not',data = df,palette='Set1').set(title="Fertility Preference")
```

We can see that the number of people who refused far outweighs the number of people who said yes. This is important to remember as we start thinking about how to build our model.

We’ll do a couple more visualizations:

```python
sns.countplot(x='wanted_child_later_or_not', hue ='religious_influence', data=df,palette='Set1').set(title="Fertility Preference & Women Empowered")
```

```python
#Removing religious influence since it does not has much influence on our dependent variable
df.drop('religious_influence',axis=1,inplace=True)
```

## Data Preparation: Converting Categorical Features

At this stage, we need to change the categorical variables to a format that our Linear Regression model will understand. We can do this by converting categorical features to dummy variables. First, we’ll double-check which features are categorical:

```python
df.info()
```

Now, we’ll convert all “non-null object” columns to dummy variables using the following process:

```python
df.columns
```

```python
#dumming all categorical variables
#
women_education = pd.get_dummies(df['women_education'],drop_first=True)
engaged_activity_for_income = pd.get_dummies(df['engaged_activity_for_income'],drop_first=True)

help_in_family_business_past_week = pd.get_dummies(df['help_in_family_business_past_week'],drop_first=True)

estimation_overall_happiness = pd.get_dummies(df['estimation_overall_happiness'],drop_first=True)

area = pd.get_dummies(df['area'],drop_first=True)

father_education = pd.get_dummies(df['father_education'],drop_first=True)

#drop columns not needed
#dropping : all columns being replaced with dummies
df.drop(['women_education','engaged_activity_for_income','help_in_family_business_past_week','estimation_overall_happiness','area','father_education'],axis=1,inplace=True)
#df = pd.concat([df,women_education,engaged_activity_for_income,help_in_family_business_past_week,estimation_overall_happiness,area,father_education,religious_influence],axis=1)
```

```python
df = pd.concat([df,women_education,engaged_activity_for_income,help_in_family_business_past_week,estimation_overall_happiness,area,father_education],axis=1)
```

```python
df.columns
```

```python
df.columns = ['wanted_child_later', 'age_at_marriage', 'wanted_child_later_or_not',
       'Middle', 'None/Preschool', 'Primary', 'Secondary', 'engaged_activity_for_income', 'help_in_family_business_past_week',
       'SOMEWHAT HAPPY', 'SOMEWHAT UNHAPPY', 'VERY HAPPY', 'VERY UNHAPPY',
       'area', 'Middle', 'None/Preschool', 'Primary', 'Secondary']
```

```python
df.head(100)
```

Finally, we need to convert our labels and drop any unhelpful columns. Notice that in the df.drop() stage of the code above we dropped the ‘religious_influence’ column since we concluded during preliminary visualizations that it did not provide enough information to be helpful to the model.

```python
#need to get labels
labels = pd.DataFrame(df['wanted_child_later_or_not'])
labels.value_counts()
```

```python
labels.wanted_child_later_or_not[labels.wanted_child_later_or_not == 'YES'] =1
labels.wanted_child_later_or_not[labels.wanted_child_later_or_not == 'NO'] = 0

df = df.drop(['wanted_child_later_or_not','wanted_child_later'],axis=1)
```

```python
df.head(100)
```

```python
df = df.apply(pd.to_numeric)
labels=labels.apply(pd.to_numeric)
```

## Data Preparation: Train, Test, Split

We will now split our data into training data sets and test data sets:

```python
X_train, X_test, y_train,y_test = train_test_split(df,labels,test_size=0.3,random_state=0)
```

## Building the Model

Now that we have our training and test data sets, we can train our model and make predictions:

```python
model = LogisticRegression(class_weight='balanced')

model.fit(X_train,y_train)
prediction_log = model.predict(X_test)
```

As you can see, the argument “class_weight = ‘balanced’” was passed to the model. This is due to what we noticed in our very first data visualization — that the number of people who no was much larger than the number of people who agreed.

## Model Evaluation

```python
print(classification_report(y_test,prediction_log))
```

We can see that our model has a <b >95%</b> precision rate for women refusals and a 39% precision rate for women agreed to have more childern. This is no doubt a result of our data imbalance.

## Feature Analysis

To better understand the performance of our model, we can investigate each individual feature. To do so, we’ll start by getting each individual feature’s coefficient score:

```python
importance = model.coef_.flatten()
```

And then plot each feature’s score:

```python
pyplot.rcParams['figure.figsize'] =(10,10)
pyplot.barh(df.columns,importance,color='g')
pyplot.title('Bar plot summary of feature importance')
pyplot.xlabel('score')
pyplot.show()
```

Scores marked with a zero coefficient, or very near zero coefficient, indicate that the model found those features unimportant and essentially removed them from the model. Positive scores indicate a feature that predicts class 1 (“yes”). Negative scores indicate a feature that predicts class 0 (“no”).
